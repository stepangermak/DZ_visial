```python
#динамика составляющей индекса по регионам
dynam_reg = df[['Region', 
              'Economy (GDP per Capita)',
              'Family',
              'Health (Life Expectancy)',
              'Freedom',
              'Trust (Government Corruption)',
              'Generosity']].groupby('Region').mean()
dynam_reg
```

```python
dynam_reg.plot(kind='bar', stacked=True)
```
